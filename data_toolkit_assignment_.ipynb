1. What is NumPy, and why is it widely used in Python?

- It is widely used in Python for it fast computing and ease in handling of large data sets which , if done in Python lists, is quite difficult.

2. How does broadcasting work in NumPy?
-  Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding their dimensions to match each other.
For example ,


```python
#broadcasting example>.
import numpy as np
arr = np.zeros((3,3), dtype = int)
arr
```

```python
#adding a scalar to arr without writing any explicit loop
arr + 5
# The scalar 10 is broadcasted to each element
```
